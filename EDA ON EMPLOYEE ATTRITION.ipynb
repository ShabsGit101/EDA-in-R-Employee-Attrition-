```r
# Importing the dataset
HRData <- read.csv("Employee-Attrition.csv")
```

```r
#Exploring the structure and columns of the dataset. 
str(HRData)      
head(HRData)
```

```r
is.na(HRData)
```

```r
HRData <- HRData %>% select(-EmployeeCount,-EmployeeNumber,-Over18,-StandardHours,-MaritalStatus)
```

```r
library(ggplot2)

y <- table(HRData$Attrition)
colors <- c("green", "red")
labels <- c("No", "Yes")
pie(y, labels = labels, col = colors, main = "Attrition", clockwise = TRUE,)
```

```r
ggplot(HRData, aes(x=ï..Age)) + geom_density()
```

```r
ggplot(HRData, aes(x=Department, fill=Attrition)) + 
  geom_bar(position="dodge") +
  labs(title="Count Plot of Attrition by Department")
```

```r
ggplot(HRData, aes(x=Education, y=MonthlyIncome)) + 
  geom_point() + 
  geom_smooth(method="lm", se=FALSE) + 
  labs(x="Education", y="MonthlyIncome") + 
  theme_bw()
```

```r
ggplot(HRData, aes(x=ï..Age, fill=Attrition)) + 
  geom_bar() + 
  theme_minimal() + 
  labs(title="Attrition by Age", x="ï..Age", y="Count") + 
  theme(plot.title = element_text(hjust = 0.5), legend.position="bottom") + 
  guides(fill=guide_legend(title="Attrition"))
```
